## Dataset Overview

Inspecting the dataset structure, data types, and basic information to understand its contents.

```python
# Dataset information
print("Dataset Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)

print("\nDataset Information:")
df.info()
```

## Handling Missing Values

Checking for missing values and applying appropriate handling techniques such as imputation or removal.

```python
# Check for missing values and removing if there
print("Missing Values:")
print(df.isnull().sum())

df = df.dropna()

print("\nDataset Shape After Handling Missing Values:")
print(df.shape)
```

## Outlier Detection and Removal

Using Interquartile Range (IQR) method to identify and remove extreme values from numerical features.

```python
import numpy as np

numeric_cols = df.select_dtypes(include=np.number).columns

Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) |
            (df[numeric_cols] > (Q3 + 1.5 * IQR)))

print("Number of Outliers per Column:")
print(outliers.sum())

df_clean = df[~outliers.any(axis=1)].copy()

print("\nShape Before Outlier Removal:", df.shape)
print("Shape After Outlier Removal:", df_clean.shape)
```

## Encoding Categorical Data

Converting categorical species labels into numerical values using Label Encoding.

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df_clean['species_encoded'] = encoder.fit_transform(df_clean['species'])

print("Class Mapping:")
for i, cls in enumerate(encoder.classes_):
    print(f"{cls} = {i}")

df_clean.head()
```

## Feature Standardization

Standardizing numerical features so that they have a mean of 0 and a standard deviation of 1.

```python
from sklearn.preprocessing import StandardScaler

features = ['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']

sch = StandardScaler()

df_clean[features] = sch.fit_transform(df_clean[features])

df_clean.head()
```
