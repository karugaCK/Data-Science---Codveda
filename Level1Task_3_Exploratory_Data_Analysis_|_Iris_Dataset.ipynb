# Exploratory Data Analysis (EDA) - Iris Dataset

In this notebook, I performs Exploratory Data Analysis to understand the structure, distribution, and relationships within the Iris dataset using statistical summaries and visualizations.

```python

!pip install -q gdown

import gdown
import pandas as pd

file_id = "1dp9LlrNHkXEFwn1osjQ3J0HygYWCkEnO"

gdown.download(
    f"https://drive.google.com/uc?id={file_id}",
    "dataset.csv",
    quiet=False
)

df = pd.read_csv("dataset.csv")
df.head()
```

## Dataset Overview

Now to nspect the structure, data types, and basic information about the dataset.

```python
print("Shape:", df.shape)
print("\nInfo:")
df.info()

print("\nFirst 5 rows:")
df.head()
```

## Summary Statistics

Computing key statistical measures such as mean, median, variance, and standard deviation.

```python
df.describe()
```

```python
print("Median Values:")
print(df.median(numeric_only=True))

print("\nVariance Values:")
print(df.var(numeric_only=True))
```

## Data Visualization of Distributions

By use of histograms to understand the distribution of each numerical feature.

```python
import matplotlib.pyplot as plt

df.hist(figsize=(15,8))
plt.suptitle("Feature Distributions", fontsize=14)
plt.show()
```

## Relationship Analysis using Scatter Plots

Now I visualize relationships between features using scatter plots.

```python
plt.figure(figsize=(8,4))
plt.scatter(df['sepal_length'], df['sepal_width'])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['petal_length'], df['petal_width'])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width")
plt.show()
```

## Box Plots for Outlier Detection

Box plots help visualize spread and detect outliers in numerical features.

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Box Plots of Numerical Features")
plt.show()
```

## Correlation Matrix

Now I analyze relationships between numerical features using a correlation matrix.

```python
plt.figure(figsize=(8,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

## Insights from EDA

### These are my key Observations:
- Petal length and petal width are highly correlated.
- Setosa species is easily separable from others based on petal features.
- Sepal features show weaker correlation compared to petal features.
- Most distributions appear fairly balanced with no extreme skewness.
